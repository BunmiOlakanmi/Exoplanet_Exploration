# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna() 
df.head()
```

```python
# Select features (columns). This will also be used as your x values.
selected_X = df[['koi_period',  'koi_time0bk', 'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_slogg', 'koi_srad', 'ra', 'dec', 'koi_kepmag']]
selected_X.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# X= df.drop(columns=['koi_disposition'])
# X
```

```python
X = selected_X
X
```

```python
y = df["koi_disposition"]
target_names = ['CONFIRMED', 'FALSE POSITIVE', 'CANDIDATE']
```

## Pre-processing

```python
# Split data to train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train
```

```python
# Scale your data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Train a Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf = rf.fit(X_train_scaled, y_train)
rf.score(X_test_scaled, y_test)
```

```python
print(f"Training Data Score: {rf.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {rf.score(X_test_scaled, y_test)}")
```

```python
# Get important variables in the RF
feature_names = X_train.columns
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
# Calculate classification report - RF
from sklearn.metrics import classification_report
predictions = rf.predict(X_test_scaled)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```python
from sklearn.model_selection import GridSearchCV
rf_cv = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)
rf_cv.fit(X_train_scaled, y_train)
```

```python
print(rf_cv.best_params_)
print(rf_cv.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions2= rf_cv.predict(X_test_scaled)
```

```python
print('Test Acc: %.3f' % rf_cv.score(X_test_scaled, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions2,
                            target_names))
```
